```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import mplhep as hep 
import uproot
from scipy.stats import bootstrap

plt.style.use(hep.style.CMS)
```

```python
dat=pd.read_csv("EmissionSpectrum_LSO_Ce_295K.csv",header=None)
plt.plot(dat[0],dat[1],"o-")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Relative Intensity (a.u.)")
plt.title("LSO:Ce Emission Spectrum")
plt.grid()
plt.show()
```

```python
dat=pd.read_csv("AbsorptionLength_LSO_Ce.txt",header=None,delim_whitespace=True)
plt.plot(dat[0],dat[1],"o-")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Absorption Length (cm)")
plt.title("LSO:Ce Absorption Length")
plt.grid()
plt.show()
```
